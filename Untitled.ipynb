```python
import re

# Fonction pour supprimer la partie entre crochets
def remove_brackets(value):
    if isinstance(value, str):
        return re.sub(r"\s*\[.*?\]\s*", "", value).strip()
    return value

# Appliquer la fonction à la colonne 'indicator_value'
df_filtered['indicator_value'] =df_filtered['indicator_value'].apply(remove_brackets)
```

```python
df_filtered['indicator_value'] = pd.to_numeric(df_filtered['indicator_value'], errors='coerce')
```
